sorting - 데이터를 순서대로 정리하는 것

```python
import pandas as pd
import numpy as np
```

```python
table1 = [
    ['2019-01-05',1000,'False','Gum'],
    ['2019-01-03',3000,'True',"snack"],
    ['2019-01-06',2000,'True',"beverage"],
    ['2019-01-06',1000,'True',"Gum"],
]
pd.DataFrame(table1)
```

```python
pd.DataFrame(table1,columns=['일자','가격','구매여부','제품'])
```

```python
colname = ['일자','가격','구매여부','제품']
df1 = pd.DataFrame(table1,columns=colname,index = ['A','B','C','D'])
df1
```

```python
df1.sort_index()
```

```python
df1.sort_index(axis=1)  #축을 1로 잡아주면 index순서가 아니라 column 순서대로 정리 할 수 있음
```

```python
df1.sort_index(axis=1,ascending = False)    #뒤에서 부터 가져옴 내림차순으로
```

```python
df1.sort_values(by='일자')    #날짜 순으로 가져올 수 있다
```

```python
df1.sort_values(by='일자',ascending=False).head(2)    #내림차순 2개
```

```python
df1.sort_values(by=['가격','일자'],ascending=[False,True])  #가격은 내림차순(가격 ->False) , 일자는 오름차순 (일자 ->True)
```
