```python
import pandas as pd
import numpy as np
```

```python
table1 = [
    ['2019-01-05',1000,'False','Gum'],
    ['2019-01-03',3000,'True',"snack"],
    ['2019-01-06',2000,'True',"beverage"],
    ['2019-01-06',1000,'True',"Gum"],
]
pd.DataFrame(table1)
```

```python
pd.DataFrame(table1,columns=['일자','가격','구매여부','제품'])
```

```python
colname = ['일자','가격','구매여부','제품']
df1 = pd.DataFrame(table1,columns=colname,index = ['A','B','C','D'])
df1
```

- 특정 위치에 있는 데이터를 찾아올 때 사용

```python
df1['제품']
```

```python
df1[['제품','일자']]
```

- location -> 특정 위치에 있는 데이터를 가져온다

```python
df1.loc['B']
```

```python
df1.loc['A':'D']   #A~D까지 가져옴
```

```python
df1['가격']['B']   #값만 가져올 수 있음
```

```python
df1.iloc[0:2]       #위치기반 location   /  loc 항목기반 location
```

```python
#loc는 항목을 써줘야되고 , iloc는 matarix형태로 인지 해서 0~2데이터를 가져온다
```

```python
df1.iloc[0,2]    #0,2항목에 대한 데이터
```

- 비교연산을 이용한 Indexing

```python
df1['가격'] > 2000
```

```python
df1[df1['가격'] > 2000]    #2000원 이상의 데이터만 데이터 프레임 형태로 가져옴
```

```python
df1[df1['제품'] == 'Gum']
```

```python
df1[(df1['제품'] == 'Gum')&(df1['구매여부']=='True')]
```

- and : &

- or : | shift + \

```python
condition1 = (df1['제품']=='Gum')
condition2 = (df1['구매여부']=='True')
df1[condition1|condition2]
```

```python
df1[condition1&condition2]
```

```python
df1[df1['제품']!='Gum']
```

```python
df1['제품'].isin(['snack'])        #isin => 제품이라는 데이터 안에 snack 이 있냐 없냐
```

```python
df1[df1['제품'].isin(['snack', 'Gum'])]     #  == 표시를 계속 쓰기 힘드니까 isin을 사용함
```

```python
df1[~df1['제품'].isin(['snack', 'Gum'])]   
#snack이랑 Gum이 없는 자료를 가져오고 싶다 -->  '~' 사용!!
#but and 조건 이용할 때 주의해야함 and도 or로 뒤집힘
```
