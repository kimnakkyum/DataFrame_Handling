```python
import pandas as pd
import numpy as np
```

```python
table1 = [
    ['2019-01-05',1000,'False','Gum'],
    ['2019-01-03',3000,'True',"snack"],
    ['2019-01-06',2000,'True',"beverage"],
    ['2019-01-06',1000,'True',"Gum"],
]
pd.DataFrame(table1)
```

```python
pd.DataFrame(table1,columns=['일자','가격','구매여부','제품'])
```

```python
colname = ['일자','가격','구매여부','제품']
df1 = pd.DataFrame(table1,columns=colname,index = ['A','B','C','D'])
df1
```

```python
df1.mean()     #가격밖에 연속형자료가 없기 때문에 가격만 나옴
```

```python
df1['가격'].mean()
```

```python
df1.min()
```

```python
df1.max()
```

```python
df1['가격'].max()
```

```python
df1.describe() #DataFrame 형태 #연속형 자료에 대한 모든 요약 정리를 해줌 / 갯수, 평균 , 표준편차, 최솟값,25%에 있는 데이터 , ...
```

```python
df1['가격'].describe()    #Series 형태
```

```python
df1['제품'].describe()   #연속형 자료가 아니어도 (명목형 자료) 요약 정리를 해줌
```

```python
df1['제품','일자','구매여부'].describe()  #여러개로 할 때 Series로 입력하면 에러가 남
```

```python
df1[['제품','일자','구매여부']].describe()   #DataFrame 으로 입력을 해주어야 함
```

```python
obj1 = ['제품','일자','구매여부']
df1[obj1].describe()    #이렇게하면 편하게 사용 가능
```
