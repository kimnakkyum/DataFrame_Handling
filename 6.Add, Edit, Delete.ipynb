```python
import pandas as pd
import numpy as np
```

```python
table1 = [
    ['2019-01-05',1000,'False','Gum'],
    ['2019-01-03',3000,'True',"snack"],
    ['2019-01-06',2000,'True',"beverage"],
    ['2019-01-06',1000,'True',"Gum"],
]
pd.DataFrame(table1)
```

```python
pd.DataFrame(table1,columns=['일자','가격','구매여부','제품'])
```

```python
colname = ['일자','가격','구매여부','제품']
df1 = pd.DataFrame(table1,columns=colname,index = ['A','B','C','D'])
df1
```

```python
df1['환불여부'] = '정상'
df1
```

```python
df1['환불여부'] = ['정상','환불','환불','정상']   #데이터 갯수 맞춰주지 않으면 에러남
df1
```

```python
condition1 = (df1['구매여부'] == 'True') & (df1['환불여부']=='환불')
df1.loc[condition1, '판매여부'] = False   #loc함수를 이용해서 새로운 DataFrame을 만들어줌 
df1.loc[~condition1, '판매여부'] = True
df1
```

```python
df1.info()   #판매여부에 missing data가 2개밖에 없다
```

```python
df1['가격(clean)']= df1['가격']*2
df1
```

```python
df1['가격(clean)']=df1['가격'].mean()
df1
```

```python
df2 = df1.drop('가격(clean)',axis=1)     
#drop : 기존의 데이터를 index나 column중심으로 삭제 
df2                                      
#하지만 아직 df1에는 '가격(clean)'이 살아있음     
#drop 은 그냥 눈에서만 안보이게 해주는 것
```

```python
df1.drop('가격(clean)',axis=1, inplace = True)    #inplace = True -> 진짜 column 자체에서 삭제를 시켜줌 inplace  (잘 안씀)
```

```python
df1.drop(['가격(clean)','판매여부'].axis=1.inplace=True)   #가격이 이미 삭제돼서 오류로 뜸
```

```python
df1['환불여부(clean)'] = df1['환불여부'].replace('정상',1).replace('환불',0)
df1
```
