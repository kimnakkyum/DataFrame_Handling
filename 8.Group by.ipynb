- Pivot_Talbe : trend를 파악할 수 있음

```python
pd.pivot_table(data = abalone, index='sex',values = 'shell_weight', aggfunc='sum')  #DataFrame으로 이쁘게 뽑을 수 있음
```

```python
result1 = pd.pivot_table(data = abalone, index=['sex','length'],values = 'shell_weight', aggfunc='sum')  #sex별 length별 shell_weight 총합
result1
```

```python
result2 = pd.pivot_table(data = abalone, index=['sex','length'],values = 'shell_weight', aggfunc='sum')  #sex별 length별 shell_weight 총합
result2.reset_index().head(15)
```

```python
result3 = pd.pivot_table(data = abalone, index=['sex','length'],values = 'shell_weight', aggfunc='sum')  #sex별 length별 shell_weight 총합
result3.reset_index().sort_values(by='shell_weight',ascending = False).head(3)                #python의 장점이 다 들어가있음
```
